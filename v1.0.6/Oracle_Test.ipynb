```python
import cx_Oracle
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import io

# Oracle Instant Client 경로 지정 (필요 시)
# cx_Oracle.init_oracle_client(lib_dir=r"H:\Project\08_clavicle_software_python\sw\v1.0.6\instantclient-basic-windows.x64-23.4.0.24.05\instantclient_23_4")

# 데이터베이스 연결 설정
dsn_tns = cx_Oracle.makedsn('1.253.96.148', 1521, service_name='kitech')
try:
    connection = cx_Oracle.connect(user='SYSTEM', password='kitech', dsn=dsn_tns)
    print("Connection successful!")
except cx_Oracle.DatabaseError as e:
    print(f"Database connection error: {e}")
    exit()

cursor = connection.cursor()

def download_image(image_id, output_path):
    try:
        sql = "SELECT image_data FROM image_table WHERE id = :id"
        cursor.execute(sql, id=image_id)
        result = cursor.fetchone()
        
        if result:
            lob = result[0]
            # LOB 데이터 읽기
            image_data = lob.read()
            print(f"Image data length: {len(image_data)} bytes")
            
            with open(output_path, 'wb') as file:
                file.write(image_data)
            print("Image downloaded successfully!")

            img = mpimg.imread(io.BytesIO(image_data))
            plt.figure()
            plt.imshow(img)
            plt.axis('off')  # 축을 숨깁니다.
            plt.show()
        else:
            print("No data found for the provided ID.")
    except Exception as e:
        print(f"Error downloading image: {e}")

# 이미지 다운로드 호출
download_image(1, r'H:\Project\08_clavicle_software_python\sw\v1.0.6\downloaded_image2.png')

# 연결 종료
cursor.close()
connection.close()
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLineEdit, QPushButton, QVBoxLayout, QLabel, QMessageBox
import cx_Oracle

class DBConnectionWindow(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()
    
    def initUI(self):
        self.setWindowTitle('Oracle DB 접속 정보 입력')
        self.setGeometry(300, 300, 400, 200)
        
        layout = QVBoxLayout()
        
        self.hostLabel = QLabel('호스트 (IP 주소 또는 도메인):')
        self.hostInput = QLineEdit()
        layout.addWidget(self.hostLabel)
        layout.addWidget(self.hostInput)
        
        self.portLabel = QLabel('포트:')
        self.portInput = QLineEdit()
        layout.addWidget(self.portLabel)
        layout.addWidget(self.portInput)
        
        self.sidLabel = QLabel('SID:')
        self.sidInput = QLineEdit()
        layout.addWidget(self.sidLabel)
        layout.addWidget(self.sidInput)
        
        self.userLabel = QLabel('사용자 이름:')
        self.userInput = QLineEdit()
        layout.addWidget(self.userLabel)
        layout.addWidget(self.userInput)
        
        self.passwordLabel = QLabel('비밀번호:')
        self.passwordInput = QLineEdit()
        self.passwordInput.setEchoMode(QLineEdit.Password)
        layout.addWidget(self.passwordLabel)
        layout.addWidget(self.passwordInput)
        
        self.connectButton = QPushButton('연결')
        self.connectButton.clicked.connect(self.connectToDB)
        layout.addWidget(self.connectButton)
        
        self.setLayout(layout)
    
    def connectToDB(self):
        host = self.hostInput.text()
        port = self.portInput.text()
        sid = self.sidInput.text()
        user = self.userInput.text()
        password = self.passwordInput.text()
        
        dsn = cx_Oracle.makedsn(host, port, sid=sid)
        try:
            connection = cx_Oracle.connect(user=user, password=password, dsn=dsn)
            QMessageBox.information(self, '연결 성공', '데이터베이스에 성공적으로 연결되었습니다.')
            connection.close()
        except cx_Oracle.DatabaseError as e:
            QMessageBox.critical(self, '연결 실패', f'데이터베이스 연결에 실패했습니다: {e}')

def main():
    app = QApplication(sys.argv)
    window = DBConnectionWindow()
    window.show()
    sys.exit(app.exec_())

if __name__ == '__main__':
    main()
```
